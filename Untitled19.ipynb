```python
median_building_age = data['building_age'].median()
data['building_age'].fillna(median_building_age, inplace=True)
median_year_built = data['year_built'].median()
data['year_built'].fillna(median_year_built, inplace=True)
median_building_size_m2 = data['building_size_m2'].median()
data['building_size_m2'].fillna(median_building_size_m2, inplace=True)
```

```python
data_filled = data.fillna(0)
data['property_condition'].fillna('Tidak Diketahui', inplace=True)
data['building_orientation'].fillna('Tidak Diketahui', inplace=True)
```

```python
data_encoded = pd.get_dummies(data, columns=['property_type', 'property_condition', 'building_orientation'])
print(data_encoded)
```

```python
data_numeric = data.select_dtypes(include=['number'])
print(data_numeric)
```

```python
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data_numeric)
print(data_scaled)
```

```python
data.isnull().sum()
```

```python
X = [col for col in data_encoded.columns if 'property_type_' in col] + ['building_size_m2']
y = ['price_in_rp']
df_encoded = pd.get_dummies(data_encoded, columns=['building_size_m2'], drop_first=True)
print(df_encoded)
```

```python
X_train, X_test, y_train, y_test = train_test_split(data_encoded[X], data_encoded[y], test_size=0.2, random_state=42)
print(f'Untuk pelatihan (X), {X_train.shape}')
print(f'Untuk pengujian (X), {X_test.shape}')
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
plt.scatter(X_train['building_size_m2'], y_train)
plt.xlabel('building_size_m2')
plt.ylabel('price_in_rp')
```

```python
y_prediksi = model.predict(X_test)
print(y_prediksi)
```
